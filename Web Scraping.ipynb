```python
temp=set ()
base_url="https://quotes.toscrape.com/page/{}/"
for j in range(1,11):
    changing_url=base_url.format(j)
    out=requests.get(changing_url)
    soup=bs4.BeautifulSoup(out.text,'lxml')
    comp=soup.select(".author")
    for i in comp:
        temp.add(i.text)
```

```python
"""List of authors with the no of pages known"""
temp
```

```python
quotes=soup.select ('.text') 
```

```python
"""list of quotes """
for i in quotes:
    print(i.text)
```

```python
"""List of authors with no of pages unknown """
valid=True 
writer=set()
page=1
base_url="https://quotes.toscrape.com/page/{}/"
while valid:
    url=base_url.format(page)
    res=requests.get(url)

    if 'No quotes found!' in res.text:
        valid=False

    soup=bs4.BeautifulSoup(res.text,'lxml')
    for name in soup.select('.author'):
        writer.add(name.text)
    
    page+=page
```
